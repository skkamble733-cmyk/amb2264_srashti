```python
import matplotlib.pyplot as plt
labels=["rainfall","temperature","humidity","sunlight","radiations","frost"]
sizes=[33,52,12,17,42,48]
separated=[.1,0,0,0,0,0]
plt.pie(sizes,labels=labels,autopct="%1.1f%%",explode=separated)
plt.title("environmental factors affecting crop growth")
plt.show()
```

```python
import matplotlib.pyplot as plt
slices=[6,8,5,4]
activities=["tillering","panicle emergence","flowering","milking stage"]
cols=["c","m","r","b"]
plt.pie(slices,
        labels=activities,
        colors=cols,
        startangle=90,
        shadow=True,
        explode=(0,0.1,0,0),
        autopct="%1.1f%%")
plt.title("critical stages of water in RICE")
plt.show()
```

```python
years=[2010,2012,2014,2016,2018,2020]
open_sensex=[10,8,30,45,50,50.5]
close_sensex=[13,5,34,38,50,50]

plt.plot(years,open_sensex,"--",label ="open",marker="o",markerfacecolor="green")
plt.plot(years,close_sensex,label = "close")
plt.ylabel("productivity(tons/year)")
plt.xlabel("years")
plt.title("productivity of crop trend:bajra")
plt.grid(True)
plt.show()
```

```python
import numpy as np
years=[2010,2012,2014,2016,2018,2020]
open_sensex=[10,8,30,45,50,50.5]
close_sensex=[13,5,34,38,50,50]
ind=np.arange(len(years))
print(ind)
width=0.3
plt.bar(ind,open_sensex,width,label="open")
plt.bar(ind+width,close_sensex,width,label="close",color="r")
plt.ylabel("productivity(tons/year)")
plt.xlabel("years")
plt.title("productivity of crop trend:bajra")
plt.xticks(ind,years,rotation=70)
plt.legend()
plt.show()


```

```python
import numpy as np
years=[2010,2012,2014,2016,2018,2020]
open_sensex=[10,8,30,45,50,50.5]
close_sensex=[13,5,34,38,50,50]
ind=np.arange(len(years))
print(ind)
width=0.3
plt.barh(ind,open_sensex,width,label="open")
plt.barh(ind+width,close_sensex,width,label="close",color="g")
plt.ylabel("productivity(tons/year)")
plt.xlabel("years")
plt.title("productivity of crop trend:bajra")
plt.yticks(ind,years)
plt.legend()
plt.show()
```
