```python
num=[1,2,3,4]
n=[]
for  i in num:
  n.append(i+2)
print("new list",n)
print("original list",num)
```

```python
num=[1,2,3,4]
n=[i+2 for i in num]
print("new list",n)
print("original list",num)
```

```python
num=[1,2,3,4]
n=[]
for  i in num:
  if i%4==0:
    n.append(i)
print("num divided by 4",n)
print("original list",num)
```

```python
num=[1,2,3,4]
n=[i for i in num if i%4==0]
print("num divided by 4",n)
print("original list",num)
```

```python
equipment=["tractor","plough","harvestor","seeder","sprayer","thresher"]
cost=[850000,25000,1200000,60000,18000,95000]
updated_c=[]
for y in c:
  updated_c.append(round((y*1.08),2))
print("actual costs:",c)
print("updated costs:", updated_c)

```

```python
equipment_updated=[]
for i in range(len(equipment)):
  if updated_c[i]>1000000:
    equipment_updated.append("high investment")
  else:
      equipment_updated.append(equipment[i])
print("equipment list after replacement:\n",equipment_updated)

equipments_updated=["high investment"if c>1000000 else e for e,c in zip(equipment,cost)]
print("equipment list after replacement:\n",equipments_updated)
```

```python
categories=[]
for c in updated_c:
 if c < 1000000:
   categories.append("low")
 elif c <= 500000:
   categories.append("medium")
 else:
  categories.append("high")
print("cost categories:",categories)
```

```python
subsidy_cost = []
for c in cost:
   if c<100000:
     subsidy_cost.append(int(c*0.85))
   else:
    subsidy_cost.append(c)
print("cost after subsidy:",subsidy_cost)
subsidy_cost=[int(c*0.85) if c<100000 else e for e,c in zip(equipment,cost)]
print(" equipment cost list  after subsidy:\n",subsidy_cost)
```

```python
average_cost= sum(cost)/len(cost)
count=0
print("average cost:",average_cost)
for i in range (len(cost)):
  if cost  [i]>average_cost:
    count=count+1
print("number of equipments having more than average cost :",count)
```

```python
count1=0
for e,c in zip (equipment,cost):
  if c>average_cost:
   count1=count1+1
print(count1)

```
