```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
datasets=pd.read_csv("/content/Product_purchase (1).csv")
#importing datasets
x=datasets.iloc[:,[2,3]].values
y=datasets.iloc[:,4].values
print(x)
print(y)
print(datasets.columns[[2,3]])
print(datasets.columns[4])
```

```python
#splitting dataset into training set and test set
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)

#feature scaling
from sklearn.preprocessing import StandardScaler
sc_x=StandardScaler()
x_train=sc_x.fit_transform(x_train)
x_test=sc_x.transform(x_test)
```

```python
#decision tree
# create decision tree classifier object
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier(criterion='entropy',max_depth=None,min_samples_split=5,min_samples_leaf=2)

model.fit(x_train,y_train) #training the model
print(model)
```

```python
y_pred_dec=model.predict(x_test)   #predict the testing model OR testing the model
```

```python
from sklearn.metrics import confusion_matrix #making confusion matrix
cm=confusion_matrix(y_test,y_pred_dec)
cm
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred_dec))
```

```python
#create an object of algorithm/model
from sklearn.linear_model import LogisticRegression
modellr=LogisticRegression(random_state=0)
modellr.fit(x_train,y_train)
y_pred_dec=modellr.predict(x_test)   #predict the testing model
y_pred_dec

from sklearn.metrics import confusion_matrix #making confusion matrix
cmlr=confusion_matrix(y_test,y_pred_dec)
cmlr

from sklearn.metrics import accuracy_score
print("LogisticRegression",accuracy_score(y_test,y_pred_dec))
```

```python
#fitting the classifier into training set
from sklearn.ensemble import RandomForestClassifier
RFmodel=RandomForestClassifier(n_estimators=200,criterion="entropy",random_state=0)
RFmodel.fit(x_train,y_train)
print(RFmodel)

#predict thwe test set results
y_pred_rf=RFmodel.predict(x_test)
y_pred_rf

from sklearn.metrics import confusion_matrix  #making the confusion matrix
cmrf=confusion_matrix(y_test,y_pred_rf)
cmrf
from sklearn.metrics import accuracy_score
print("RandomForestClassifier",accuracy_score(y_test,y_pred_rf))
```
