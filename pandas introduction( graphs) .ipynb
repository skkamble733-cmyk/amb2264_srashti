```python
print(df)
```

```python
#group by year and sum production
yearly_production=df.groupby("Year")["Production_Tons"].sum()
print("\nyearly production:")
print(yearly_production)
```

```python
#SCATTER PLOT
import matplotlib.pyplot as plt
plt.scatter(df["Rainfall_mm"],df["Production_Tons"])
plt.xlabel("Rainfall(mm)")
plt.ylabel("Production(Tons)")
plt.title("Rainfall vs Production")
plt.show()
```

```python
#bar plot
plt.figure(figsize=(8,5))
plt.bar(yearly_production.index,yearly_production.values,color="skyblue")
plt.xlabel("Year")
plt.ylabel("Total Production(Tons)")
plt.title("Total Crop  Production per year")
plt.xticks(yearly_production.index)
plt.show()
```

```python
area_year=df.groupby("Year")["Area_Hectares"].sum()
area_year.plot(kind="bar",color="green")
plt.xlabel("Year")
plt.ylabel("Total Area(Hectares)")
plt.title("Total cultivated area per Year")
plt.show()
```

```python
import pandas as pd
df =pd.read_csv("agri_data.csv")
print(df)
```

```python
#plot seperately
wheat=df[df["Crop"]=="Wheat"]
rice=df[df["Crop"]=="Rice"]
plt.figure()
plt.plot(wheat["Year"],wheat["Production_Tons"],label="Wheat")
plt.plot(rice["Year"],rice["Production_Tons"],label="Rice")
plt.xlabel("Year")
plt.ylabel("Production(Tons)")
plt.title("Crop Production Over Years")
plt.legend()
plt.show()
```

```python
data_2022=df[df["Year"]==2022]
print(data_2022)
plt.figure(figsize=(6,4))
plt.bar(data_2022["Crop"],data_2022["Production_Tons"],color=["green","orange"])
plt.xlabel("Crop")
plt.ylabel("Production(Tons)")
plt.title("Crop Production in 2022")
plt.show
```
