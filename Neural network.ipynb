```python
importimport tensorflow as tf
mnist=tf.keras.datasets.mnist
(x_train,y_train),(x_test,y_test)=mnist.load_data()
```

```python
print("Train images shape:",x_train.shape)
print("Train labels shape:",y_train.shape)
print("Train images shape:",x_test.shape)
print("Train labels shape:",y_test.shape)
```

```python
import matplotlib.pyplot as plt
plt.imshow(x_train[10],cmap="gray")
plt.title(f"Label:{y_train[10]}")
plt.axis("off")
plt.show()
```

```python
plt.figure(figsize=(10,2))
for i in range(5):
  plt.subplot(1,5,i+1)
  plt.imshow(x_train[i])
  plt.title(f"Label:{y_train[i]}")
  plt.axis("off")
plt.show()
```

```python
#step2:normalise values(0 to 1)
train_images=x_train/255.0
test_images=x_test/255.0
```

```python
#step3:add channel dimension(important for CNN)
train_images=train_images.reshape
test_images=test_images.reshape
```

```python
from tensorflow.keras import layers,models
model=models.Sequential([
    layers.Conv2D(16,(3,3),activation="relu",input_shape=(28,28,1)),
    layers.MaxPooling2D(2,2),
    layers.Flatten(),
    layers.Dense(32,activation="relu"),
    layers.Dense(10,activation="softmax")])
```

```python
#step4:build CNN model
```

```python
#STEP5:compile model
model.compile(optimizer="adam",
              loss="sparse_categorical_crossentropy",
              metrics=["accuracy"])
```

```python
#step6:train model
model.fit(train_images,y_train, epochs=3)
```

```python
#step7:test model
test_loss, test_acc =model.evaluate(test_images,y_test)
print("Accuracy:",test_acc)
```
