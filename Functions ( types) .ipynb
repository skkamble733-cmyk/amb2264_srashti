```python
def fact (m):                                       # functions types,1.
   fact=1
   if (m>0):
    for i in range(1,m+1):
      fact=fact*i
    print("factorial of m is ",fact)
   else:
    print("enter a valid number:")
n=int(input("enter a number"))
fact(n)
```
