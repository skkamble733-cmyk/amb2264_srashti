```python
data={"farm":["farm a","farm b","fame c","farm d"],"crop":["wheat","rice","sorghum","maize"],"area_hectare":[10,15,8,12],"production_tons":[30,60,24,36]}
df=pd.DataFrame(data)
print(df)
df.info()
df.head()
df.tail()
df.describe()
```

```python
df["yield_ton_per_hectare"]=df["production_tons"]/df["area_hectare"]
print(df)
df.iloc[[1,3]]
print(df["crop"]=="wheat")
print()
wheat_farms=df[df["crop"]=="wheat"]
print(wheat_farms)
```

```python
high_yield=df[df["yield_ton_per_hectare"]>3]
print(high_yield)
```

```python
sorted_df=df.sort_values(by="production_tons",ascending=False)
print(sorted_df)
```

```python
import pandas as pd
data={"farm":["farm a","farm b","fame c","farm d"],"crop":["wheat","rice","sorghum","maize"],"area_hectare":[10,15,8,12],"production_tons":[30,60,24,36]}
df=pd.DataFrame(data)
avg_yield=df.groupby("crop")["yield_ton_per_hectare"].sum()
print(avg_yield)
```
