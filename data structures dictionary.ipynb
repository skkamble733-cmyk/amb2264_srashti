```python
cgpa={1:8.9,2:5.6,4:6.7,7:9.1,8:5.3}
print (cgpa.keys())
print (list(cgpa.keys()))
print(cgpa.values())
print (cgpa.items())
```

```python
keylist=list(cgpa.keys())
valuelist=list(cgpa.values())
minval=min(cgpa.values())
print("\nvalue=",minval,"key ",keylist[valuelist.index(minval)])
```

```python
maxval=max(cgpa.values())
print("\nvalue=",maxval,"key ",keylist[valuelist.index(maxval)])
```

```python
print("\n key having min value ",min(cgpa,key=cgpa.get))
print("\n key having max value ",max(cgpa,key=cgpa.get))
```

```python
print("is key 2 present ", 2 in cgpa.keys())
print("is value 5.6 present ",5.6 in cgpa.values())
print ("is key 10 preset ",10 in cgpa.keys())
```
